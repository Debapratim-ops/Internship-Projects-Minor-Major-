LIBRARIES

```python
# importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#loading of the Dataset
df = pd.read_csv('Housing.csv')
df
```

FIRST FIVE RECORDS DISPLAY

```python
# Displaying top 5 records
df.head()
```

NULL VALUES

```python
# Checking missing values from Dataset
df.isnull().sum()
```

TAKING 6 IMPORTANT COLUMNS INTO CONSIDERATION

```python
# Important Columns to take into consideration for pricing of houses
df = df[['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']]
df
```

SCATTERPLOT

```python
# Scatterplot between Area and Prices
plt.Figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'])
plt.xlabel('area')
plt.ylabel('price')
plt.title('Area V Prices')
plt.show()
```

DATASET USING FOR THE PREDICTIONS

```python
# Dataset to use for predictions:
df
```

PREDICTIONS OF HOUSE PRICING :

```python
X = df.drop('price', axis = 1)
Y = df['price']

# Splitting the dataset into training and testing sets
X_train, X_test, Y_train,Y_test = train_test_split(X, Y, test_size= 0.2, random_state= 42)

# Creating a multiple regression model
model = LinearRegression()
model

#training
model.fit(X_train, Y_train)

# Making the Predictions
Y_pred = model.predict(X_test)
Y_pred
```

PREDICTION OF A NEW HOUSE

```python
# Prediction of a new house price
New_House = [[3000, 3, 2, 2, 1]]
predicted_price = model.predict(New_House)
print("Predicted Price for the New House is :", predicted_price[0])
```
